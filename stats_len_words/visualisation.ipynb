```python
import pickle
import matplotlib.pyplot as plt

dict = pickle.load(open("dict.pickle", "rb"))
```

```python
for k, v in dict.items():
    print(k)
```

```python
bins = list(e for e in range(17))
for k, v in dict.items():
    plt.hist(dict[k], bins=bins)
    plt.gca().set(title='CER distribution for words of length ' + str(k), ylabel='Frenquence')
    plt.show()
```

```python
dict2 = {}
for k, v in dict.items():
    dict2[k] = [e for e in v if e!=0]
```

```python
bins = list(e for e in range(17))
for k, v in dict2.items():
    plt.hist(dict2[k], bins=bins)
    plt.gca().set(title='CER distribution for words of length ' + str(k), ylabel='Frenquence')
    plt.show()
```

```python
moyennes = []
for k, v in dict.items():
    try:
        moyenne = int((sum(v)/len(v))/k*100)
        #print(moyenne)
        moyennes.append(moyenne)
    except ZeroDivisionError:
        moyenne = 0

y = list(e for e in range(1, len(moyennes)))
plt.scatter(y, moyennes[1:]) #, bins=bins)
print(y)
print(moyennes)
```
